```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

from sklearn.cluster import KMeans

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
df = pd.read_csv("Cricket.csv", sep=",", encoding="ISO-8859-1", header=0)
df.head()
```

```python
cluster_df = df[["Ave", "SR"]]
cluster_df.head()
```

```python
ss = StandardScaler()
cluster_df = ss.fit_transform(cluster_df)
cluster_df.shape
```

```python
kmeans4 = KMeans(n_clusters=4, random_state=100)
kmeans4.fit(cluster_df)
```

```python
df["KMeans4_Label"] = kmeans4.labels_
df[df["Player"].str.contains("Kohli")]
```

```python
# complete linkage
plt.figure(figsize=(20,12))
mergings = linkage(cluster_df, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
cluster_labels = cut_tree(mergings, n_clusters=4).reshape(-1, )
cluster_labels
```

```python
df["HC_labels"] = cluster_labels
df.head()
```
